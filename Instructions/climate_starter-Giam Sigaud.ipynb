# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model  
# from 10-Advanced-Data-Storage-and-Retrieval/2/Activities/05-Ins_Reflection
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
#from 10-Advanced-Data-Storage-and-Retrieval/2/Activities/05-Ins_Reflection
Base.classes.keys()
```

```python
# Inspecting measurement Table columns
# 10-Advanced-Data-Storage-and-Retrieval/2/Activities/07-Ins_Exploration/Solved/Ins_Inspector
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Inspecting station Table columns
# 10-Advanced-Data-Storage-and-Retrieval/2/Activities/07-Ins_Exploration/Solved/Ins_Inspector
inspector = inspect(engine)
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Save references to each table
#from 10-Advanced-Data-Storage-and-Retrieval/2/Activities/05-Ins_Reflection
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
#from 10-Advanced-Data-Storage-and-Retrieval/2/Activities/05-Ins_Reflection
session = Session(engine)
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station).count()
print(f"Station Count: {station_count}")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_count = (session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all())
station_count
```

```python
# Using the station id from the previous query, 
most_active_station = 'USC00519281'

# calculate the lowest temperature recorded,
# Reminder: TOBS = temperature observation data
lowest_temp = (session.query(Measurement.tobs)
               .filter(Measurement.station == most_active_station)
               .order_by(Measurement.tobs.asc())
               .first())
lowest_temp = lowest_temp[0]
print(f"The lowest temperature recorded was {lowest_temp} degrees Farenheit.")

# highest temperature recorded, and 
highest_temp = (session.query(Measurement.tobs)
                .filter(Measurement.station == most_active_station)
                .order_by(Measurement.tobs.desc())
                .first())
highest_temp = highest_temp[0]
print(f"The highest temperature recorded was {highest_temp} degrees Farenheit.")

#average temperature of the most active station?
avg_temp = (session.query(func.avg(Measurement.tobs))
                  .filter(Measurement.station == most_active_station))
avg_temp = '{0:.3}'.format(avg_temp[0][0])
print(f"The average temperature recorded was {avg_temp} degrees Farenheit.")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station 
temp_observations = session.query(Measurement.date, Measurement.tobs).filter(Measurement.station == most_active_station).filter(func.strftime("%Y-%m-%d", Measurement.date) >= '2016-08-23').all()

#save as a data frame
temp_observations_df = pd.DataFrame(temp_observations, columns = ['date', 'temperature'])
temp_observations_df.set_index('date', inplace = True)
temp_observations_df.head()
```

```python
# and plot the results as a histogram
fig, ax = plt.subplots()
temp_observations_df.plot.hist(bins = 10, ax = ax)

#set labels
ax.set_xlabel('Temperature (F)')
ax.set_ylabel('Frequency')
ax.set_title("Temps (F) '16-'17 from Top HI Observation Station")

#save figure
plt.savefig("Images/temperature_histo.png")

#plot
plt.tight_layout()
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
